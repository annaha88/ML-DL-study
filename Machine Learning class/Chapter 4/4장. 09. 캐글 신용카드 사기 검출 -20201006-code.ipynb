```python
# 반복 사용 위해, copy function 만들기. 원본을 훼손하지 않고, 새로운 카피본으로 작업 하기 위함

# 인자로 입력받은 DataFrame을 복사한 뒤, 불필요한 'Time' 컬럼 삭제시킨 후 반환
def get_preprocessed_df(df=None):
    df_copy = df.copy()
    df_copy.drop('Time', axis = 1, inplace =True)
    return df_copy
```

```python
# 학습/테스트 데이터 분리 함수 생성
from sklearn.model_selection import train_test_split

def get_train_test_dataset(df=None):
    # 파라미터 => df : 원본 데이터프레임
    df_copy = get_preprocessed_df(df)
    
    X_feature  = df_copy.iloc[:,:-1]
    y_labels = df_copy.iloc[:, -1]
    
    # 학습/테스트 데이터 분리 : 레이블이 동일 분포로 분리되어야 함
    X_train, X_test, y_train, y_test = train_test_split(
        X_feature, y_labels, test_size=0.3, random_state=0, stratify=y_labels
    )
    
    return X_train, X_test, y_train, y_test
```

stratify(계층화) 하는 이유
- 극도로 불균형한 데이터셋인 경우, 
- 레이블을 기준으로 계층형으로 나누어야 함.
- 즉 답이 1이거나 0인애들이 데이터 세트에 고르게 분포되도록 데이터를 나누어야 함.

ex) train 세트에 0,1이 있는 비율과 비슷하게 test 세트에도 들어가도록 만들기. 


```python
# function 이용해서 데이터셋 분리
X_train, X_test, y_train, y_test = get_train_test_dataset(card_df)
```

학습데이터와 테스트 데이터 내의 1 비율이 0.17% 정도로 비슷하게 맞춰졌다 

```python
# 검증하는 함수 선언 (오차행렬, 정확도, 정밀도, 재현율, auc점수)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_auc_score
```

```python
def get_clf_eval(y_test, pred, pred_prob):
    confu = confusion_matrix(y_test, pred)
    precision = precision_score(y_test, pred)
    accu = accuracy_score(y_test, pred)
    recall = recall_score(y_test, pred)
    # auc 여기 뭔가 잘못된듯. pred에 pred_proba 가 들어가는게 맞을듯? 근데 오류는 나지 않았었다..(0, 1값을 확률로 인식한다면 그렇긴 하네)
    auc = roc_auc_score(y_test, pred_prob)
    
    print('오차행렬')
    print(confu)
    out_str = '정확도: {0:.4f}, 정밀도:{1:.4f}, 재현율: {2:.4f}, AUC: {3:.4f}'
    print(out_str.format(accu, precision, recall, auc))
```

#### Ligth GBM 모델 생성

```python
# 학습시킬 함수 만들기 : 파라미터=> 모델, 학습데이터, 테스트 데이터와 답
def get_model_train_eval(model, ftr_train=None, ftr_test=None, tgt_train=None, tgt_test=None):
    model.fit(ftr_train, tgt_train)  # 학습
    pred= model.predict(ftr_test)  # 예측값 추출
    pred_prob = model.predict_proba(ftr_test)[:,1]
    get_clf_eval(tgt_test, pred, pred_prob)   # 성능평가
    
```

Class 와 상대적으로 관련이 높은 컬럼들 : V14,  V17

여기서는 V14 컬럼의 이상치만 찾아서 제거할 것이다

```python
# 이상치를 가지는 row의 인덱스 검출 함수 생성
def get_outliner(df=None, col=None, weight=1.5):
    fraud=df[df['Class']==1][col]
    q_25 = np.percentile(fraud.values, 25)
    q_75 = np.percentile(fraud.values, 75)
    iqr = q_75-q_25  # IQR(Inter Quantile Range) 범위지정
    iqr_weight = iqr*weight
    low_value = q_25 -iqr_weight    # 이상치 기준이 되는 최소값
    high_value = q_75 + iqr_weight   # 이상치 기준이 될 최대값
    
    out_index = fraud[(fraud<low_value)|(fraud > high_value)].index
    return out_index
    
```

```python
out_index_14 = get_outliner(df=card_df, col='V14')  # weight 는 디폴트 처리되있어서 생략가능
print('V14 이상치 데이터 인덱스 : ', out_index_14)
```

```python
out_index = get_outliner(df=card_df, col='V17')  # weight 는 디폴트 처리되있어서 생략가능
print('V17 이상치 데이터 인덱스 : ', out_index)
```

```python
# 카피 함수 수정하기, => 이상치 제거 코드 추가

def get_preprocessed_df(df=None):
    df_copy = df.copy()
    amount_n = np.log1p(df_copy['Amount'])
    df_copy.insert(0, 'Amount_Scaled', amount_n)
    
    df_copy.drop(['Time','Amount'], axis = 1, inplace =True)
    
    # 이상치 처리 코드
    out_index=get_outliner(df=df_copy, col='V14')
    df_copy.drop(out_index, axis=0, inplace=True)
    
    return df_copy
```

### 데이터 전처리 방법 (극도로 불균형한 데이터셋)
   1. StandardScaler : 원래 값을 정규분포값으로 변화 (값 사이의 분포 변화가 없음) => 극도로 불균형한 데이터셋에서 의미없을수있다.
   2. log1p : 로그스케일로 변환, 값 분포가 고르지 않은 경우 유용함
   3. 이상치 처리: 튀는 데이터 삭제
   4. 오버 샘플링 : 전체 데이터 대비, 비율이 매우 작은 유의미한 데이터 개수를 임의로 늘려 학습시킴.

### 머신러닝/딥러닝
- 기본 작동방식 : 예측값(대표값=>default:평균)과 원래값(대표값 => default:평균)차이를 최소화하도록 학습하는 방식
- 이 대표값에 영향을 크게 미치는 인자들은 전처리 해주어야함.(이상치 삭제 등)
- 예측값과 원래값과의 차이 = 오차(Loss) == 오류(Error)
